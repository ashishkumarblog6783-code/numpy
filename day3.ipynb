```python
original_n = int(input("enter number"))
n = original_n
rev = 0
while(n > 0):
  rem = n % 10  # Correctly get the last digit
  rev = (10 * rev) + rem
  n = n // 10 # Use integer division to remove the last digit
if(rev == original_n):
    print("palindrome")
else:
    print("not a palindrome")
```

```python
original = int(input("enter the number"))
num = original
rev = 0
while(num > 0):
  rem = num % 10  # Get the last digit
  rev = (rev * 10) + rem # Build the reversed number
  num = num // 10 # Remove the last digit using integer division

if(original == rev):
  print("palindrome")
else:
  print("not a palindrome")
```
