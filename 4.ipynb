iteration


```python
import numpy as np
var=np.array([1,2,4,5,6,7,9])
for i in var:
  print(i)
# The inner loop 'for l in k:' is not needed for a 1D array as 'k' itself is an element.
# If you intend to iterate over elements of a multi-dimensional array, you would need to adjust the array structure or indexing.
# For this 1D array, the first loop 'for i in var:' already achieves iteration over elements.
```

```python
var=np.array([[[3,4,5,7],[4,1,1,6]]])
for i in np.nditer(var):
  print(i)
```

```python
var=np.array([[[3,4,5,7],[4,1,1,6]]])
for i,d in np.ndenumerate(var):
  print(i,d)
```

copy vs view


```python
import numpy as np
var=np.array([1,2,3,4])
co=var.copy()
print(var)
print(co)
```

```python
x=np.array([2,4,6,8,9,0])
vi=x.view()
print(x)
print(vi)
```

join array


```python
var=np.array([2,4,5,7])
var1=np.array([3,5,6,7])
ar=np.concatenate((var,var1))
print(ar)
```

```python
v=np.array([[1,2],[3,4]])
b=np.array([[3,1],[4,4]])
ar=np.concatenate((v,b),axis=1)
print(ar)
```

```python
v=np.array([1,2,3,4])
b=np.array([3,1,4,4])
ar=np.stack((v,b),axis=1)
ar1=np.hstack((v,b))  #row
ar2=np.vstack((v,b))  #column
ar3=np.dstack((v,b))  #height
print(ar)
print(ar1)
print(ar2)
print(ar3)
#
```
