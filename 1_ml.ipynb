```python
import matplotlib.pyplot as plt
x=[20,30,57,80]
y=["c","python","java","r"]
ex=[0.3,0.7,0.0,0.3]
plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%", shadow=True,radius=1.5,labeldistance=1.2,startangle=0,counterclock=False,)

plt.show()
```

```python
import matplotlib.pyplot as plt
x=[20,30,57,80]
y=["c","python","java","r"]

plt.stem(x,y,linefmt=":",markerfmt="ro",bottom=0,orientation="horizontal")
plt.show()
```
