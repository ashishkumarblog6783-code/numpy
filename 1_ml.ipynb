```python
import matplotlib.pyplot as plt
x=['mon','tue','wed','thu','fri']
y=[13,56,32,87,44]
plt.plot(x,y)
plt.title('bakery sales this week')
plt.xlabel('days of week')
plt.ylabel('sales per day')

plt.show()
```

```python
import matplotlib.pyplot as plt
month=[1,2,3,4]
sales=[1000,2000,4500,6789]
plt.xlabel('month')
plt.ylabel('sales')
plt.title('2026 sales data')
plt.grid(color='pink')
plt.plot(month,sales,color='blue',linestyle='--',linewidth=2,marker='o',label='2026 sales data')
plt.show()
```

```python
import matplotlib.pyplot as plt
product=['a','b','c','d']
sales=[1000,3400,5432,6788]
plt.bar(product,sales,color='orange',label='produt sales')
plt.xlabel('product')
plt.ylabel('sales')
plt.title('product sales comparison')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
regions=['north','south','east','west']
revenue=['2000','4322','32122','34232']
plt.pie(revenue,labels=regions,autopct='%1.1%%',colors=['gold','skyblue','lightgreen','violet'])
plt.title('revenue')
plt.show()
```

```python
import matplotlib.pyplot as plt
scores=[12,34,5,6,22,55,66,12,55,90,9,66,45,32]
plt.hist(scores,bins=5,color='blue',edgecolor='black')
plt.xlabel('scores')
plt.ylabel('number os students')
plt.title('scores distribution')
plt.show()
```

```python
import matplotlib.pyplot as plt
hours_studies=[1,2,3,5,7,8,9]
exam_scores=[20,34,55,78,90,54,55]
plt.scatter(hours_studies,exam_scores,color='green',marker='o',label='syudent data')
plt.xlabel('hours of study')
plt.ylabel('exam scores')
plt.title('relation between study and hours')
plt.grid()
```
