```python
import numpy as np
var=np.array([2,3,4,5,6,7])
print(np.min(var),np.argmin(var))
print(np.max(var),np.argmax(var))
```

```python
var=np.array([[2,5,7,9],[7,4,0,9]])
print(np.min(var,axis=1))
```

```python
var=np.array([4,7,0])
print(np.sin(var))
print(np.cos(var))
```

```python
var=np.array([7,9,33])
print(np.cumsum(var))
```

```python
var=np.array([[1,5],[8,0]])
print(var)
print(var.shape)
print(var.reshape)
```
